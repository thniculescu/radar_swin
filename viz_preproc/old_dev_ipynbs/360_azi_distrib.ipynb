```python
# %matplotlib inline
from nuscenes.nuscenes import NuScenes
import matplotlib.pyplot as plt
import pprint
import numpy as np
from pyquaternion import Quaternion
from nuscenes.utils.data_classes import RadarPointCloud



nusc = NuScenes(version='v1.0-mini', dataroot='/imec/other/dl4ms/nicule52/datasets/nuscenes/v1.0-mini/', verbose=True)
nusc.list_scenes()
```

```python
from nuscenes.utils.geometry_utils import transform_matrix, view_points

maxdist = 0
maxRCS = 0
minRCS = 1e6
minmaxangs = {}

HIST_PLOTS = False
PLOTS = True
STATS = False
LIMS = 60
RadarPointCloud.default_filters()
# RadarPointCloud.disable_filters()

all_points = np.empty((3, 0))
angle_bins = np.zeros(360)

for scene_id in range(len(nusc.scene)):
# for scene_id in range(1):

    scene_angle_bins = np.zeros(360)
    print(nusc.scene[scene_id]['name'])
    print(nusc.scene[scene_id]['nbr_samples'])
    cur_sample = nusc.get('sample', nusc.scene[scene_id]['first_sample_token'])
    # for nr_samps in range(nusc.scene[scene_id]['nbr_samples'] - 1):
    for nr_samps in range(3):
        # pprint.pp(cur_sample)
        #count sweeps to the next sapmle
        samp_sweep_nr_points = 0
        ref_chan = 'LIDAR_TOP'
        ref_sd_record = nusc.get('sample_data', cur_sample['data'][ref_chan])

        if PLOTS:
            fig, ax = plt.subplots(1, 2, figsize=(16, 8))
        #set xlim and ylim for both plots
        
        # ax[0].grid(True)
        
        for chan_type, chan_tok in cur_sample['data'].items():
            if 'RADAR' in chan_type:
                pc, times = RadarPointCloud.from_file_multisweep(nusc, cur_sample, chan_type, ref_chan, nsweeps=6)

                # Transform radar velocities (x is front, y is left), as these are not transformed when loading the
                # point cloud.
                sd_record = nusc.get('sample_data', chan_tok)
                # pprint.pp(sd_record)
                radar_cs_record = nusc.get('calibrated_sensor', sd_record['calibrated_sensor_token'])
                ref_cs_record = nusc.get('calibrated_sensor', ref_sd_record['calibrated_sensor_token'])
                velocities = pc.points[8:10, :]  # Compensated velocity
                velocities = np.vstack((velocities, np.zeros(pc.points.shape[1])))
                velocities = np.dot(Quaternion(radar_cs_record['rotation']).rotation_matrix, velocities)
                velocities = np.dot(Quaternion(ref_cs_record['rotation']).rotation_matrix.T, velocities)
                velocities[2, :] = np.zeros(pc.points.shape[1])

                cs_record = nusc.get('calibrated_sensor', ref_sd_record['calibrated_sensor_token'])
                pose_record = nusc.get('ego_pose', ref_sd_record['ego_pose_token'])
                ref_to_ego = transform_matrix(translation=cs_record['translation'],
                                              rotation=Quaternion(cs_record["rotation"]))

                # Compute rotation between 3D vehicle pose and "flat" vehicle pose (parallel to global z plane).
                ego_yaw = Quaternion(pose_record['rotation']).yaw_pitch_roll[0]
                rotation_vehicle_flat_from_vehicle = np.dot(
                    Quaternion(scalar=np.cos(ego_yaw / 2), vector=[0, 0, np.sin(ego_yaw / 2)]).rotation_matrix,
                    Quaternion(pose_record['rotation']).inverse.rotation_matrix)
                vehicle_flat_from_vehicle = np.eye(4)
                vehicle_flat_from_vehicle[:3, :3] = rotation_vehicle_flat_from_vehicle
                viewpoint = np.dot(vehicle_flat_from_vehicle, ref_to_ego)
                # print(pc.points.shape)

                my_pc = view_points(pc.points[:3, :], viewpoint, normalize=False)
                my_pc = my_pc.T

                if my_pc.shape[0] == 0:
                    continue

                # print(my_pc.shape)
                xs = my_pc[:, 0]
                ys = my_pc[:, 1]

                #convert to range and angle in 360 degrees
                ranges = np.hypot(xs, ys)
                angles = np.degrees(np.arctan2(ys, xs))
                angles[angles < 0] += 360
                speeds = np.hypot(velocities[0], velocities[1])
                rcs = pc.points[5, :].T

                samp_sweep_nr_points += my_pc.shape[0]


                
                # mask angles, only when range is < 100
                mask = ranges < 60
                ranges = ranges[mask]
                angles = angles[mask]
                speeds = speeds[mask]
                rcs = rcs[mask]

                print('SHAPES:', ranges.shape, angles.shape, speeds.shape, rcs.shape)
                # print(rcs)

                scene_angle_bins += np.histogram(angles, bins=360, range=(0, 360))[0]

                #append angles and ranges to all_points
                all_points = np.append(all_points, np.vstack((angles, ranges, speeds)), axis=1)

                if STATS:
                    # print(nr_samps, chan_type, xs.shape, ys.shape)
                    maxdist = max(maxdist, np.max(np.hypot(xs, ys)))
                    if chan_type not in minmaxangs:
                        minmaxangs[chan_type] = (4, -4)
                    else:
                        if 'BACK' in chan_type:
                            minmaxangs[chan_type] = (min(minmaxangs[chan_type][0], np.min(np.arctan2(-ys, -xs))), max(minmaxangs[chan_type][1], np.max(np.arctan2(-ys, -xs))))
                        else:
                            minmaxangs[chan_type] = (min(minmaxangs[chan_type][0], np.min(np.arctan2( ys,  xs))), max(minmaxangs[chan_type][1], np.max(np.arctan2( ys,  xs))))

               
                if PLOTS:
                    ax[0].plot(my_pc[:, 0], my_pc[:, 1], '.', markersize=3)
                    nusc.render_sample_data(chan_tok, ax=ax[1], nsweeps=1, axes_limit=LIMS, underlay_map=False, with_anns=True, verbose=False)
                    ax[0].set_xlim(-LIMS, LIMS)
                    ax[0].set_ylim(-LIMS, LIMS)
                    # ax[1].set_xlim(-LIMS, LIMS)
                    # ax[1].set_ylim(-LIMS, LIMS)
                    ax[0].grid()
                    # ax[1].axis(xmin=-LIMS, xmax=LIMS, ymin=-LIMS, ymax=LIMS)
                    ax[1].axis('on')
                    ax[1].grid()
                    ax[0].set_xticks(np.arange(-LIMS, LIMS, 10))
                    ax[0].set_yticks(np.arange(-LIMS, LIMS, 10))
                    ax[1].set_xticks(np.arange(-LIMS, LIMS, 10))
                    ax[1].set_yticks(np.arange(-LIMS, LIMS, 10))
                    ax[0].set_title('Manual Radar Point Cloud')
                    ax[0].set_aspect('equal')
                    ax[0].plot(0, 0, 'x', color='red')

        # print('Nr samp:', nr_samps,'Samp_sweep points:', samp_sweep_nr_points)

        if PLOTS:
            plt.show()

        cur_sample = nusc.get('sample', cur_sample['next'])

    angle_bins += scene_angle_bins

    # if HIST_PLOTS:
    #     fig, ax = plt.subplots(1, 1, figsize=(16, 8))
    #     ax.plot(np.arange(360), scene_angle_bins)
    #     ax.set_title('Scene Angle Histogram')
    #     ax.set_xlabel('Angle (degrees)')
    #     ax.set_ylabel('Count')
    #     ax.grid()
    #     plt.show()


if HIST_PLOTS:
    fig, ax = plt.subplots(1, 1, figsize=(16, 8))
    #plot angle bins so that front direction is in the center, and not at the edge
    angle_bins = np.roll(np.flip(angle_bins), 180)
    ax.plot(np.arange(-180, 180, 1), angle_bins)
    ax.set_title('Total Angle Histogram')
    ax.set_xlabel('Angle (degrees) clockwise from front')
    ax.set_ylabel('Count')
    ax.grid()
    plt.show()

    #make a heatmap from all_points, for the range distribution for each angle
    # fig, ax = plt.subplots(1, 1, figsize=(16, 8))
    # ax.hist2d(all_points[0], all_points[1], bins=(360, 100), range=[[0, 360], [0, 100]])
    # ax.set_title('Range distribution for each angle')


if STATS:
    print(maxdist)
    pprint.pp(minmaxangs)
    minmaxangs['RADAR_BACK_LEFT'] = (np.pi + minmaxangs['RADAR_BACK_LEFT'][0], -np.pi + minmaxangs['RADAR_BACK_LEFT'][1])
    minmaxangs['RADAR_BACK_RIGHT'] = (np.pi + minmaxangs['RADAR_BACK_RIGHT'][0], -np.pi + minmaxangs['RADAR_BACK_RIGHT'][1])
    for k, v in minmaxangs.items():
        print(k, np.degrees(v[0]), np.degrees(v[1]))
```
